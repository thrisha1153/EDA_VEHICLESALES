Average New Sales over Years

```python
new=df.groupby("Year ")["Total Sales New"].mean()
```

```python
plt.figure(figsize=(10,6))

# Creating subplots
plt.subplot(2, 1, 1)

# Plotting the bar plot
sns.barplot(x=new.index, y=new.values)
plt.xlabel('Year')
plt.ylabel('Total Sales New')
plt.title('Average Total Sales New by Year')

# Creating subplots
plt.subplot(2, 1, 2)

# Plotting the line plot
plt.plot(new.index, new.values, marker='o', color='red', linewidth=2)
plt.xlabel('Year')
plt.ylabel('Total Sales New')
plt.title('Trend of Average Total Sales New by Year')

plt.tight_layout()  # Adjusts the layout to prevent overlap
plt.show()
```

There was a normal increase in the average sales from 2003 to 2005 but after 2005 there is a gradual decrease in the sales with the minmum sales in the year 2009. After 2010 there is a gradual increase increase in the sales from 2010  onwards. The maximum sales is in the year 2023.

AVERAGE USED CAR SALES

```python
new2=df.groupby("Year ")["Total Sales Used"].mean()
```

```python
plt.figure(figsize=(10,6))

# Creating subplots
plt.subplot(2, 1, 1)

# Plotting the bar plot
sns.barplot(x=new2.index, y=new2.values)
plt.xlabel('Year')
plt.ylabel('Total Sales New')
plt.title('Average Total Sales New by Year')

# Creating subplots
plt.subplot(2, 1, 2)

# Plotting the line plot
plt.plot(new2.index, new2.values, marker='o', color='red', linewidth=2)
plt.xlabel('Year')
plt.ylabel('Total Sales New')
plt.title('Trend of Average Total Sales New by Year')

plt.tight_layout()  # Adjusts the layout to prevent overlap
plt.show()
```

```python
plt.figure(figsize=(10,6))
x=new.index
plt.plot(x, new.values,marker='o', color='green', linewidth=2,label="new")
plt.plot(x,new2.values, marker='o', color='red', linewidth=2,label="used")  # Add a line plot
plt.xticks(rotation='vertical')
plt.legend()
plt.show()
```
